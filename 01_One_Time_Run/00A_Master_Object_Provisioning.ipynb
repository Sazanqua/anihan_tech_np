```python
spark.sql(f"create catalog if not exists anihan_tech_{env}")
```

```python
schemas = [
    "00_master_00_lz",
    "00_master_01_brz",
    "00_master_02_slv",
    "00_master_03_gld"
]

for schema in schemas:
    spark.sql(f"create schema if not exists anihan_tech_{env}.{schema}")
```

```python
volumes = [
    "lz_vl", 
    "brz_vl", 
    "slv_vl", 
    "gld_vl"
]

for schema, volume in zip(schemas, volumes):
    spark.sql(f"create volume if not exists anihan_tech_{env}.{schema}.{volume}")
```
